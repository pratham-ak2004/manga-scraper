```python
import os
import zipfile
# from pdf2image import convert_from_path
import fitz  # PyMuPDF

def pdf_to_cbz(pdf_path, output_cbz):
    # Create temp directory
    temp_dir = "temp_pages"
    os.makedirs(temp_dir, exist_ok=True)

    print("Converting PDF pages to images...")
    doc = fitz.open(pdf_path)
    for i, page in enumerate(doc):
        pix = page.get_pixmap()
        img_path = os.path.join(temp_dir, f"{i:04d}.jpg")
        pix.save(img_path)

    # for i, page in enumerate(pages):
    #     img_path = os.path.join(temp_dir, f"{i:04d}.jpg")
    #     page.save(img_path, "JPEG")

    print("Creating CBZ file...")
    with zipfile.ZipFile(output_cbz, 'w', zipfile.ZIP_DEFLATED) as cbz:
        for img_name in sorted(os.listdir(temp_dir)):
            cbz.write(os.path.join(temp_dir, img_name), img_name)

    # Cleanup
    for f in os.listdir(temp_dir):
        os.remove(os.path.join(temp_dir, f))
    os.rmdir(temp_dir)

    print(f"Done! Saved as: {output_cbz}")
```

```python
pdf_dir = "F:\\mangas\\Vinland Saga"
out_dir = "F:\\mangas\\Vinland Saga\\cbz"
anime_name = "Vinland Saga"
```

```python
pdfs = [f for f in os.listdir(pdf_dir) if f.lower().endswith('.pdf')]

for pdf_file in pdfs:
    vol_number = ''.join(filter(str.isdigit, pdf_file))
    output_cbz = os.path.join(out_dir, f"{anime_name} Vol {vol_number}.cbz")
    pdf_path = os.path.join(pdf_dir, pdf_file)
    pdf_to_cbz(pdf_path, output_cbz)
```
